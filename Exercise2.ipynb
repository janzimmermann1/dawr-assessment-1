## Task 2: Pandas

Select all elements in the series bigger than the mean of the series itself

```python
#Input
ser1 = pd.Series([1, 2, 3, 4, 5])

#Solution
# Calculate the mean of the series
mean_value = ser1.mean()

# Select elements greater than the mean
result = ser1[ser1 > mean_value]

print(result)
```

Find all elements appearing both in ser1 and ser2 using isin() and create a new series with those elements

```python
# Input
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

#Solution
mask = ser1.isin(ser2)
result = ser1[mask]

print(result)
```

Find the canton with the highest 'Arbeitslosenquote' and print the name of that canton

Which column correlates the most with the column 'Arbeitslosenquote'? Use the pearson correlation. 

Also, if unclear: Find out what the pearson correlation actually is.

```python
# The Pearson correlation coefficient is a measure of the linear relationship between two variables. It ranges from -1 to 1, where:

# 1 indicates a perfect positive linear relationship,
# -1 indicates a perfect negative linear relationship, and
# 0 indicates no linear relationship.

# Exclude 'Arbeitslosenquote' column from the correlations
correlations = df.drop('Arbeitslosenquote', axis=1).corrwith(df['Arbeitslosenquote'])

# Find the column with the maximum correlation
max_correlation_column = correlations.idxmax()

# Get the actual correlation value
max_correlation_value = correlations.loc[max_correlation_column]

# Print the result
print("Column with the highest Pearson correlation (excluding 'Arbeitslosenquote'):", max_correlation_column)
print("Correlation value:", max_correlation_value)
```
